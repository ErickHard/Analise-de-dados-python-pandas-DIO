```python
#Criando nosso DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a Receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
#Qual o custo Total?
round(df["Custo"].sum(),2)
```

```python
#Temos a receita e custo e o total, podemos achar o lucro total
#Vamos criar uma coluna de Lucro que será Receita - Custo 
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
#Total lucro
round(df["Lucro"].sum(),2)
```

```python
#Quanto tempo que levou para o produto fosse enviado?
df["Tempo envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
#Extraindo apneas os dias
df["Tempo envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#verificando o tipo da coluna tempo de envio
df["Tempo envio"].dtypes
```

```python
#media do tempo de envio por marca
df.groupby("Marca")["Tempo envio"].mean()
```

```python
##Missing Values
```

```python
#Verificando Valores faltantes
df.isnull().sum()
```

```python
#lucro por ano e marca
#agrupando por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
 pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Grafico do total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produtos");
```

```python
#lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro X Ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head(5)
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro X Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro")
```

```python
df_2009.groupby(df_2009["Marca"])["Lucro"].sum().plot.bar(title="Lucro X Marca")
plt.xlabel("Mês")
plt.ylabel("Marca", rotation='horizontal')
```

```python
df_2009.groupby(df_2009["Classe"])["Lucro"].sum().plot.bar(title="Classe X Mês")
plt.xlabel("Mês")
plt.ylabel("Classe")
```
